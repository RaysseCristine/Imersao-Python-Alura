Aula 2 - Importação

```python
import pandas as pd
import plotly.express as px
```

```python
df_principal = pd.read_excel("/content/Tabela de ações.xlsx",sheet_name="Principal")
df_principal
```

```python
df_total_acoes = pd.read_excel("/content/Tabela de ações.xlsx",sheet_name="Total_de_acoes")
df_total_acoes
```

```python
df_ticker = pd.read_excel("/content/Tabela de ações.xlsx",sheet_name="Ticker")
df_ticker
```

```python
df_principal = df_principal[['Ativo', 'Data', 'Último (R$)', 'Var. Dia (%)']].copy()
df_principal
```

```python
df_principal = df_principal.rename(columns={'Último (R$)':'Valor_final', 'Var. Dia (%)':'var_dia_pct'}).copy()
df_principal
```

```python
df_principal['Var_pct'] = df_principal['var_dia_pct'] / 100
df_principal['Valor_inicial'] = df_principal['Valor_final'] / (df_principal['Var_pct'] + 1)
df_principal
```

```python
df_principal = df_principal.merge(df_total_acoes, left_on='Ativo', right_on='Código', how='left')
df_principal
```

```python
df_principal = df_principal.drop(columns=['Código'])
df_principal
```

```python
df_principal['Variacao_rs'] = (df_principal['Valor_final'] - df_principal['Valor_inicial'])*df_principal['Qtde. Teórica']
df_principal
```

```python
pd.options.display.float_format = '{:.2f}'.format
```

```python
df_principal['Qtde. Teórica'] = df_principal['Qtde. Teórica'].astype(int)
df_principal
```

```python
df_principal = df_principal.rename(columns={'Qtde. Teórica':'Qtd_teorica'}).copy()
df_principal
```

```python
df_principal['Resultado'] = df_principal['Variacao_rs'].apply(lambda x: 'Subiu' if x > 0 else ('Desceu' if x < 0 else 'Estável'))
df_principal
```

```python
df_principal = df_principal.merge(df_ticker, left_on='Ativo', right_on='Ticker', how='left')
df_principal = df_principal.drop(columns=['Ticker'])
df_principal
```

```python
df_chatgpt = pd.read_excel("/content/Tabela de ações.xlsx",sheet_name="ChatGPT")
df_chatgpt

df_principal = df_principal.merge(df_chatgpt, left_on="Nome", right_on=" Nome da empresa", how="left")
df_principal = df_principal.drop(columns=[" Nome da empresa"])
df_principal
```

```python
df_principal['Cat_Idade'] = df_principal['Idade (anos)'].apply(lambda x: 'Mais de 100' if x > 100 else ('Menos de 50' if x < 50 else 'Entre 50 e 100 '))
df_principal
```

```python
# Calcule a média, o máximo e o mínimo da coluna 'Variacao_rs'
media_total = df_principal['Variacao_rs'].mean()
maior = df_principal['Variacao_rs'].max()
menor = df_principal['Variacao_rs'].min()

# Calcule a média da coluna 'Variacao_rs' para os casos em que 'Resultado' é 'Subiu'
media_subiu = df_principal[df_principal['Resultado'] == 'Subiu']['Variacao_rs'].mean()

# Calcule a média da coluna 'Variacao_rs' para os casos em que 'Resultado' é 'Desceu'
media_desceu = df_principal[df_principal['Resultado'] == 'Desceu']['Variacao_rs'].mean()

# Exiba os resultados
print(f"Maior\tR$ {maior:,.2f}")
print(f"Menor\tR$ {menor:,.2f}")
print(f"Média\tR$ {media_total:,.2f}")
print(f"Média de quem subiu\tR$ {media_subiu:,.2f}")
print(f"Média de quem desceu\tR$ {media_desceu:,.2f}")
```

```python
df_principal_subiu = df_principal[df_principal['Resultado'] == 'Subiu']
df_principal_subiu
```

```python
df_analise_segmento = df_principal_subiu.groupby('Segmento')['Variacao_rs'].sum().reset_index()
df_analise_segmento
```

```python
df_analise_saldo = df_principal.groupby('Resultado')['Variacao_rs'].sum().reset_index()
df_analise_saldo
```

```python
fig = px.bar(df_analise_saldo, x='Resultado', y='Variacao_rs', text='Variacao_rs', title='Variação Reais por Resultado')
fig.show()
```

```python
# GroupBy da categoria de idades
df_analise_cat = df_principal.groupby('Cat_Idade').agg(
    {'Variacao_rs': 'sum', 'Cat_Idade': 'count'}).rename(
    columns={'Variacao_rs': 'Soma da Variação', 'Cat_Idade': 'Qtd de empresas'}).reset_index()

# Resultado
print(df_analise_cat)
```

```python
# Convertendo os valores da coluna 'Soma da Variação' para inteiros
df_analise_cat['Soma da Variação'] = df_analise_cat['Soma da Variação'].astype(float).apply(lambda x: f'R$ {x:,.2f}' if x > 0 else f'-R$ {abs(x):,.2f}')
```

```python
fig = px.bar(df_analise_cat, x='Cat_Idade', y='Soma da Variação', text='Soma da Variação', title='Qtd. de Empresas por Faixa Etária', height=500, width=500)
fig.show()
```
